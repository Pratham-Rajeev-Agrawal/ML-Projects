```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
gold = pd.read_csv('/content/gld_price_data.csv')
```

```python
gold.head()
```

```python
cor = gold.drop(columns = 'Date', axis = 1).corr()
```

```python
plt.figure(figsize = (10, 10))
sns.heatmap(cor, cbar = True, square = True, annot=True, fmt='.2f', annot_kws={'size':8}, cmap = 'Blues')
```

```python
print(cor['GLD'])
```

```python
sns.distplot(gold['GLD'], color = 'green')
```

```python
x = gold.drop(columns = ['Date', 'GLD'], axis=1)
y = gold['GLD']
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
regressor = RandomForestRegressor(n_estimators = 100)
```

```python
regressor.fit(x_train, y_train)
```

```python
prediction = regressor.predict(x_test)
```

```python
error_score = metrics.r2_score(y_test, prediction)
print(error_score)
```

```python
y_test = list(y_test)
```

```python
plt.plot(y_test, color = 'blue', label = 'Actual Value')
plt.plot(prediction, color = 'green', label = 'Predicted Value')
plt.title('Actual Price VS Predicted Price')
plt.xlabel('Number Of Values')
plt.ylabel('Gold Price')
plt.legend()
plt.show()
```
