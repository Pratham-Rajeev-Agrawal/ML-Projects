```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price)
```

```python
dataframe = pd.DataFrame(house_price.data, columns = house_price.feature_names)
```

```python
print(dataframe)
```

```python
dataframe.head()
```

```python
dataframe['price'] = house_price.target
```

```python
dataframe.head()
```

```python
dataframe.shape
```

```python
dataframe.describe()
```

```python
dataframe.isnull().sum()
```

```python
correl = dataframe.corr()
```

```python
print(correl)
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(correl, cbar=True, square = True, fmt = '.1f', annot = True, annot_kws={'size':8}, cmap = 'Blues')
```

```python
x = dataframe.drop(columns='price', axis = 1)
y = dataframe['price']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 1, test_size = 0.2)
```

```python
print(x_train, x_test, y_train, y_test)
```

```python
model = XGBRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
print(prediction)
```

```python
score1 = metrics.r2_score(y_test, prediction)
```

```python
print(score1)
```

```python
score2 = metrics.mean_absolute_error(y_test, prediction)
```

```python
print(score2)
```

```python
plt.scatter(y_test, prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual prive vs predicted price")
plt.show()
```
