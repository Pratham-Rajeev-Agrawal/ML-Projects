```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
credit = pd.read_csv('/content/creditcard.csv')
```

```python
credit.head()
```

```python
credit.shape
```

```python
credit.isnull().sum()
```

```python
credit.describe()
```

```python
credit['Class'].value_counts()
```

```python
legit = credit[credit.Class == 0]
fraud = credit[credit.Class == 1]
```

```python
print(legit.shape, fraud.shape)
```

```python
legit.describe()
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n = 492)
```

```python
credit_card = pd.concat([legit_sample, fraud], axis = 0)
```

```python
credit_card.head()
```

```python
credit_card['Class'].value_counts()
```

```python
x = credit_card.drop(columns = 'Class', axis = 1)
y = credit_card['Class']
```

```python
x_train, t_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2, stratify = y)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predict = model.predict(t_test)
```

```python
print(accuracy_score(predict, y_test))
```
